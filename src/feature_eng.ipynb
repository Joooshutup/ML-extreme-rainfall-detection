```python
import pandas as pd

df = pd.read_csv('flood_data_cities.csv')

print(df.head())
```

```python
cities = ["新北", "台北", "桃園", "台中", "臺南", "高雄", "嘉義", "彰化", "雲林", "屏東", "苗栗", "花蓮", "臺東", "宜蘭"]

counts = {}
for city in cities:
    counts[city] = df["authority"].str.contains(city, na=False).sum()

city_counts = pd.DataFrame(list(counts.items()), columns=["City", "Count"]).sort_values("Count", ascending=False)
print(city_counts)
```

```python
def simple_author(a):
    if pd.isna(a):
        return "未知"
    if "雲林" in a:
        return "雲林縣"
    if "嘉義" in a:
        return "嘉義縣"
    if "彰化" in a:
        return "彰化縣"
    if "台中" in a or "臺中" in a:
        return "台中市"
    if "台南" in a or "臺南" in a:
        return "台南市"
    if "高雄" in a:
        return "高雄市"
    if "桃園" in a:
        return "桃園市"
    if "新北" in a:
        return "新北市"
    if "基隆" in a:
        return "基隆縣"
    if "台北" in a or "臺北" in a:
        return "台北市"
    if "宜蘭" in a:
        return "宜蘭縣"
    if "苗栗" in a:
        return "苗栗縣"
    if "花蓮" in a:
        return "花蓮縣"
    if "屏東" in a:
        return "屏東縣"
    if "新竹" in a:
        return "新竹市"
    if "台東" in a or "臺東" in a:
        return "台東縣"
    if "水文技術組" in a or "河川分署" in a or "內政部" in a or "農田水利" in a:
        return "中央"
    return a

df["city"] = df["city"].apply(simple_author)
```

```python
df.loc[df["flood_depth_cm"] < 0, "flood_depth_cm"] = 0.0
```

```python
df = df.dropna(subset=["flood_depth_cm", "longitude", "latitude", "city"])
```

```python
df["flood_label"] = (df["flood_depth_cm"] > 0).astype(int)
```

```python
def classify(x):
    if x == 0:
        return "無"
    elif x <= 30:
        return "低"
    elif x <= 70:
        return "中"
    else:
        return "高"
df["flood_level"] = df["flood_depth_cm"].apply(classify)
```

```python
mapping = {"無": 0, "低": 1, "中": 2, "高": 3}
df["flood_label"] = df["flood_level"].map(mapping)
```

```python
df.to_csv("flood_data_cleaned.csv", index=False, encoding="utf-8-sig")
```
