Converted Json file to CSV file

```python
import json

with open("../data/C-B0025-001.json", "r", encoding="utf-8") as f:
    data = json.load(f)

print(data.keys())
```

```python
import pandas as pd

locs = data["cwaopendata"]["resources"]["resource"]["data"]["surfaceObs"]["location"]

def precip(val):
    if val is None or val == "":
        return None
    if val == "T":
        return 0.1 
    if val == "X":
        return None
    return float(val)

rows = [
    {
        "station_id": loc["station"]["StationID"],
        "station_name": loc["station"]["StationName"],
        "station_name_en": loc["station"]["StationNameEN"],
        "date": obs["Date"],
        "precip_mm": precip(obs["weatherElements"]["Precipitation"])
    }
    for loc in locs
    for obs in loc["stationObsTimes"]["stationObsTime"] 
]


df = pd.DataFrame(rows)
print(df.head(10))
df.to_csv("../data/rain.csv", index=False, encoding="utf-8-sig")
```
